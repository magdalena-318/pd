```python
import spacy
```

```python
nlp = spacy.load("pl_core_news_lg")
```

```python
text = "Patryk mieszka na Uniwersytecie Warszawskim i mieszka w Warszawie."
```

```python
doc = nlp(text)
```

```python
type(doc)
```

```python
nlp.pipeline
```

```python
for token in doc:
  print(token.text, token.lemma_, token.pos_, token.tag_, token.shape_, token.is_stop)
```

```python
# NER = Named entity Recognition
```

```python
for ent in doc.ents:
  print(ent.text, ent.label_)
```

```python
# wyświetl wyników analizy morfosyntaktycznej dla pierwszego 1000 znaków
# wyświetl wszystkie nery dla całego tekstu
# napisz kod który zlicza ile jest osób (persName) i organizacji (orgName) w tekście
```

```python
doc_1 = nlp(kk)
```

```python
doc_1_1000 = doc_1[0:999]
```

```python
# wyświetl wyników analizy morfosyntaktycznej dla pierwszego 1000 znaków
for token in doc_1_1000:
  print(token.text, token.lemma_, token.pos_, token.tag_, token.shape_, token.is_stop)
```

```python
# wyświetl wszystkie nery dla całego tekstu
for ent in doc_1.ents:
  print(ent.text, ent.label_)
```

```python
# napisz kod który zlicza ile jest osób (persName) i organizacji (orgName) w tekście
person_count = sum(1 for ent in doc_1.ents if ent.label_ == "persName")
org_count = sum(1 for ent in doc_1.ents if ent.label_ == "orgName")
```
